 # Import Required Libraries

```python
from google.colab.patches import cv2_imshow
```

# Importing Image

```python
image=cv2.imread("muzamil.png")
cv2_imshow(image)
cv2.waitKey(0) 
```

# Converting Image into Grey Scale Image.

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow( gray_image)
cv2.waitKey(0)
```

# Inverting the grey scale image

```python
inverted_image = 255- gray_image
cv2_imshow(inverted_image)
cv2.waitKey(0)
```

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0) #image smoothing 
```

# Pencil Sketch

```python
inverted_blurred = 255- blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```

# Showing Original Image and the Pencil Sketch 

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
